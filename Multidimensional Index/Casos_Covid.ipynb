```python
df_casos.info()
df_casos['Mes'] = df_casos['Mes'].astype('int64')
df_casos['Ano'] = df_casos['Ano'].astype('int64')
df_casos
```

```python
val_mes = dataset_v['Mes'] #Se guardan los valores de la columna 
val_ano = dataset_v['Ano'] #Se guardan los valores de la columna 

vect_a = [] #Se crea un vector vacío

for n_Mes,n_Ano in zip(val_mes,val_ano):
    vect_a.append(df_casos.apply(lambda x: ((x['Ano'] == n_Ano) and (x['Mes'] == n_Mes )), axis=1).sum())
print(vect_a)
dataset_v['Casos Covid'] = vect_a
dataset_v
```

```python
vect_tri_a = []

df_agr = pd.DataFrame()
df_agr['Casos Covid'] = None

c = [0,3,6,9,12,15,18,21,24,27,30,33,36,39]
i = 0
for n in dataset_v['Mes'].index:
    #print(c[i],n)
    vect_tri_a.append(dataset_v.loc[(dataset_v['Mes'].index >= c[i]) & (dataset_v['Mes'].index < c[i+1]),'Casos Covid'].sum())
    
    if n == c[i]:
        i = 1+i
df_agr['Casos Covid'] = vect_tri_a

df_agr
```
