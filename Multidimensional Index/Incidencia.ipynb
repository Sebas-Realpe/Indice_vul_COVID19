```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
```

```python
dataset_dir = 'Casos_Covid/Casos_Ciudades.csv' #Se carga el dataset
df_casos = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas
df_casos #Se imprime el dataset
```

```python
dataset_dir = 'Casos_Covid/Poblacion/Estimacion_pre.csv' #Se carga el dataset
df_pob = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas
df_pob #Se imprime el dataset
```

```python
df = df_casos.merge(df_pob, on=["Ano","Ciudad"], how = "outer")
df
```

```python
df['Incidencia'] = df['Casos Covid']/df['Total']
df
```

```python
columnas = ['Trimestre','Ano','Ciudad','Incidencia']
df_n = df[columnas]
df_n
```

```python
export_dir = 'Incidencia/Incidencia.csv'
df_n.to_csv(export_dir, index=False, header=True)
```
