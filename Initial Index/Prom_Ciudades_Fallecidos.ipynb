## Se inicia con el pre-procesamiento de la columna FA2_SEXO_FALL

```python
df_dataset = df_dataset.sort_values('FA2_SEXO_FALL') #Organiza la lista en orden ascendente
Sex_Fall = df_dataset['FA2_SEXO_FALL'].unique() #Se verifican que valores tiene la columna FA2_SEXO_FALL
print(Sex_Fall) #Se imprime
for sex_fall in Sex_Fall:
    empty_array = np.empty([len(df_dataset),1]) #Se crea un array vacío de las dimensión del dataset x 1
    empty_array[:] = np.nan #Se asigna todo el arreglo como nulo
    df_dataset[sex_fall] = empty_array #Se agregan las nuevas columnas
    s_aux = (df_dataset.FA2_SEXO_FALL == sex_fall).astype(int) #Teniendo en cuenta el valor de la variable FA2_SEXO_FALL del dataset
    #se realiza una comparación con el valor de sex_fall, si este valor coincide se guarda un 1, en caso de que no coincida
    #Se guarda un cero. 
    df_dataset[sex_fall][s_aux.index] = s_aux.values #En la respectiva columna de sexo de fallecido del dataset
    #en cada una de las filas guardadas en s_aux se asigna el valor correspondiente; 0 en caso de que en esa fila no haya 
    #cumplido la comparación de la sentencia anterior, y 1 de que la comparación fue verdadera. 

df_dataset = df_dataset.sort_index(ascending=True)
df_dataset #Se imprime el dataset
```

```python
df_dataset.rename(columns = {1: "Fall_Hombre", 2: "Fall_Mujer",9:'No informa sexo fall'}, inplace = True) #Se renombran las columnas 
df_dataset #Se imprime el dataset
```

```python
df_dataset.drop(["FA2_SEXO_FALL"],axis = "columns",inplace=True) #Se elimina la columna 
df_dataset #Se imprime el dataset
```

## Se inicia con el preprocesamiento de la columna FA3_EDAD_FALL

```python
#Se realiza el mismo procedimiento que para FA2_SEXO_FALL pero ahora para la columna FA3_EDAD_FALL
df_dataset = df_dataset.sort_values('FA3_EDAD_FALL') #Organiza la lista en orden ascendente
Edad_Fall = df_dataset['FA3_EDAD_FALL'].unique()
print(Edad_Fall)
for edad_fal in Edad_Fall:
    empty_array = np.empty([len(df_dataset),1])
    empty_array[:] = np.nan
    df_dataset[edad_fal] = empty_array
    s_aux = (df_dataset.FA3_EDAD_FALL == edad_fal).astype(int)
    df_dataset[edad_fal][s_aux.index] = s_aux.values
    
df_dataset = df_dataset.sort_index(ascending=True)
df_dataset
```

```python
#Se crean las siguientes columnas con un valor de cero para sus filas
df_dataset["Fall_P_Infancia"] = 0
df_dataset["Fall_Infancia"] = 0
df_dataset["Fall_Juv-Adoles"] = 0
df_dataset["Fall_Adultez"] = 0
df_dataset["Fall_Per_Mayor"] = 0

df_dataset
```

```python
df_dataset.rename(columns = {999:'No informa edad fall'}, inplace = True) #Se renombra la columna
df_dataset
```

```python
#Mediante ciclos for se asignan los rangos de edad a los grupos de edad creados previamente
for x in range(0,6): 
    df_dataset.loc[df_dataset[x] == 1,"Fall_P_Infancia"] = 1
for x in range(6,12):
    df_dataset.loc[df_dataset[x] == 1,"Fall_Infancia"] = 1
for x in range(12,27):
    df_dataset.loc[df_dataset[x] == 1,"Fall_Juv-Adoles"] = 1
for x in range(27,60):
    df_dataset.loc[df_dataset[x] == 1,"Fall_Adultez"] = 1
for x in range(60,122):
    df_dataset.loc[df_dataset[x] == 1,"Fall_Per_Mayor"] = 1

df_dataset #Se imprime el dataset
```

```python
#Se eliminan las columnas
for x in range(0,121): 
    df_dataset.drop([x],axis = "columns",inplace=True)
df_dataset.drop(["FA3_EDAD_FALL"],axis = "columns",inplace=True)
df_dataset #Se imprime el dataset
```

```python
#El propósito de las siguientes líneas, es revisar cuántos municipios hay.
df_agrup_fall = df_dataset.drop_duplicates(subset=['U_DPTO','U_MPIO'],keep='last', inplace=False) #Se borran todas las filas
#En las cuales los municipios están duplicados
df_agrup_fall.reset_index(drop=True, inplace=True)#Se reinicia el index, con el fin de que el número de cada fila quede ordenado
df_agrup_fall#Se imprime el dataset
```

```python
val_fall_mun = df_agrup_fall['U_MPIO'] #Se guardan los valores de la columna U_MPIO
val_fall_dpto = df_agrup_fall['U_DPTO'] #Se guardan los valores de la columna U_DPTO
vect_A = [] #Se crea un vector vacío
vect_B = [] #Se crea un vector vacío
vect_C = [] #Se crea un vector vacío
vect_D = [] #Se crea un vector vacío
vect_E = [] #Se crea un vector vacío
vect_F = [] #Se crea un vector vacío
vect_G = [] #Se crea un vector vacío
vect_H = [] #Se crea un vector vacío
vect_I = [] #Se crea un vector vacío

for n_MPIO,n_DPTO in zip(val_fall_mun,val_fall_dpto): #El ZIP se usa para empezar a asociar cada departamento con el respectivo municipio
    #Se realiza la comparacion entre el número de municipio y el de departamento, en el caso de que sean iguales
    #Se van sumando y se van guardando en un vector según la columna a la que corresponda
    vect_A.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Hombre'].sum())
    vect_B.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Mujer'].sum())
    vect_C.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'No informa sexo fall'].sum())
    vect_D.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'No informa edad fall'].sum())
    vect_E.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_P_Infancia'].sum())
    vect_F.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Infancia'].sum())
    vect_G.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Juv-Adoles'].sum())
    vect_H.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Adultez'].sum())
    vect_I.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Fall_Per_Mayor'].sum())
    


df_agrup_fall['Fall_Hombre'] = vect_A #En el nuevo dataframe, a cada columna se le asigna el total según corresponda
df_agrup_fall['Fall_Mujer'] = vect_B
df_agrup_fall['No informa sexo fall'] = vect_C
df_agrup_fall['No informa edad fall'] = vect_D
df_agrup_fall['Fall_P_Infancia'] = vect_E
df_agrup_fall['Fall_Infancia'] = vect_F
df_agrup_fall['Fall_Juv-Adoles'] = vect_G
df_agrup_fall['Fall_Adultez'] = vect_H
df_agrup_fall['Fall_Per_Mayor'] = vect_I

df_agrup_fall #Se imprime el dataset
```

```python
export_dir = 'TG2022_Datos/Departamentos_CSV/Pre_Departamentos_fall_2.csv'
df_agrup_fall.to_csv(export_dir, index=False, header=True)
```
