## Se unen los dataframes

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

# Configuración warnings
# ==============================================================================
import warnings
warnings.simplefilter(action='ignore')
```

```python
dataset_dir = 'TG2022_Datos\Departamentos_CSV\Pre_Departamentos_viv_2.csv' #Se carga el dataset de vivienda
df_dataset_viv = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas

df_dataset_viv #Se imprime el dataset
```

```python
dataset_dir = 'TG2022_Datos\Departamentos_CSV\Pre_Departamentos_hog_2.csv' #Se carga el dataset de hogares
df_dataset_hog = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas

df_dataset_hog
```

```python
#Se unen los datasets de vivienda y hogares
df_dt_viv_hog = df_dataset_viv.merge(df_dataset_hog, on=["U_MPIO","U_DPTO","Nom_DPTO"], how="left")
df_dt_viv_hog
```

```python
dataset_dir = 'TG2022_Datos\Departamentos_CSV\Pre_Departamentos_fall_2.csv' #Se carga el dataset de fallecidos
df_dataset_fall = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas

df_dataset_fall
```

```python
df_dataset_fall = df_dataset_fall.fillna(0) #Se reemplazan los datos nulos por cero
df_dataset_fall
```

```python
#Se une el dataset anterior (Vivienda y hogares) y fallecidos
#Se reemplazan los datos nulos por ceros
df_dt_viv_hog_fall = df_dt_viv_hog.merge(df_dataset_fall, on=["U_MPIO","U_DPTO","Nom_DPTO"], how="left").fillna(0)
df_dt_viv_hog_fall
```

```python
dataset_dir = 'TG2022_Datos\Departamentos_CSV\Pre_Departamentos_per_unido.csv' #Se carga el dataset de personas
df_dataset_per = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas

df_dataset_per
```

```python
#Se une el dataset anterior (Vivienda, hogares y fallecidos) y personas
df_dt_viv_hog_fall_per = df_dt_viv_hog_fall.merge(df_dataset_per, on=["U_MPIO","U_DPTO","Nom_DPTO"], how="left")
df_dt_viv_hog_fall_per
```

```python
#Se exporta el dataset obtenido
export_dir = 'TG2022_Datos/Departamentos_CSV/Dataset_FINAL_viv_hog_fall_per.csv' #Se establece la dirección de almacenamiento del dataset
df_dt_viv_hog_fall_per.to_csv(export_dir, index=False, header=True) #Se almacena el dataset
```
