## Se inicia con el pre-procesamiento de la columna P_SEXO

```python
df_dataset = df_dataset.sort_values('P_SEXO') #Organiza la lista en orden ascendente
Per_Sexo = df_dataset['P_SEXO'].unique() #Se verifican los valores que hay en la columna P_SEXO
print(Per_Sexo) #Se imprime
for per_sexo in Per_Sexo:
    empty_array = np.empty([len(df_dataset),1]) #Se crea un array vacío de las dimensión del dataset x 1
    empty_array[:] = np.nan #Se asigna todo el arreglo como nulo
    df_dataset[per_sexo] = empty_array #Se agregan las nuevas columnas
    s_aux = (df_dataset.P_SEXO == per_sexo).astype(int) #Teniendo en cuenta el valor de P_SEXO del dataset
    #se realiza una comparación con el valor de per_sexo, si este valor coincide se guarda un 1, en caso de que no coincida
    #Se guarda un cero. 
    df_dataset[per_sexo][s_aux.index] = s_aux.values #En la respectiva columna de per_sexo del dataset
    #en cada una de las filas guardadas en s_aux se asigna el valor correspondiente; 0 en caso de que en esa fila no haya 
    #cumplido la comparación de la sentencia anterior, y 1 de que la comparación fue verdadera. 

df_dataset = df_dataset.sort_index(ascending=True)
df_dataset #Se imprime el dataset
```
