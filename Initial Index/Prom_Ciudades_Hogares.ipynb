## Se inicia con el pre-procesamiento de la columna NRO_DORMITORIO

```python
df_dataset = df_dataset.sort_values('H_NRO_DORMIT') #Organiza la lista en orden ascendente
Num_dormit = df_dataset['H_NRO_DORMIT'].unique() #Se verifican los valores que puede tomar la columna H_NRO_DORMIT
print(Num_dormit) #Se imprime
for num_dor in Num_dormit: #Se realiza un ciclo for el cual tiene en cuenta el número de dormitorios
    empty_array = np.empty([len(df_dataset),1]) #Se crea un array vacío de las dimensión del dataset x 1
    empty_array[:] = np.nan #Se asigna todo el arreglo como nulo 
    df_dataset[num_dor] = empty_array #Se agregan las columnas por cada uno de los valores que hay en num_dor
    s_aux = (df_dataset.H_NRO_DORMIT == num_dor).astype(int)#Teniendo en cuenta el valor de la variable H_NRO_DORMIT del dataset
    #se realiza una comparación con el valor de num_dor, si este valor coincide se guarda un 1, en caso de que no coincida
    #Se guarda un cero. 
    df_dataset[num_dor][s_aux.index] = s_aux.values #En la respectiva columna de número de hogares del dataset
    #en cada una de las filas guardadas en s_aux se asigna el valor correspondiente; 0 en caso de que en esa fila no haya 
    #cumplido la comparación de la sentencia anterior, y 1 en caso de que la comparación haya sido verdadera. 
    

df_dataset = df_dataset.sort_index(ascending=True) #Se organiza por el índice en orden ascendente
df_dataset #Se imprime el dataset
```

```python
#El propósito de las siguientes líneas, es revisar cuántos municipios hay.
df_agrup_hog = df_dataset.drop_duplicates(subset=['U_DPTO','U_MPIO'],keep='last', inplace=False) #Se borran todas las filas
#En las cuales los municipios están duplicados
df_agrup_hog.reset_index(drop=True, inplace=True)#Se reinicia el index, con el fin de que el número de cada fila quede ordenado
df_agrup_hog#Se imprime el dataset
```

```python
val_hog_mun = df_agrup_hog['U_MPIO'] #Se guardan los valores de la columna U_MPIO
val_hog_dpto = df_agrup_hog['U_DPTO'] #Se guardan los valores de la columna U_DPTO
vect_A = [] #Se crea un vector vacío
vect_B = [] #Se crea un vector vacío
vect_C = [] #Se crea un vector vacío
vect_D = [] #Se crea un vector vacío
vect_E = [] #Se crea un vector vacío
vect_F = [] #Se crea un vector vacío
vect_G = [] #Se crea un vector vacío
vect_H = [] #Se crea un vector vacío
vect_I = [] #Se crea un vector vacío
vect_J = [] #Se crea un vector vacío
vect_K = [] #Se crea un vector vacío
vect_L = [] #Se crea un vector vacío
vect_M = [] #Se crea un vector vacío
vect_N = [] #Se crea un vector vacío
vect_O = [] #Se crea un vector vacío
for n_MPIO,n_DPTO in zip(val_hog_mun,val_hog_dpto): #El ZIP se usa para empezar a asociar cada departamento con el respectivo municipio
    #Se realiza la comparacion entre el número de municipio y el de departamento, en el caso de que sean iguales
    #Se van sumando y se van guardando en un vector según el tipo de vivienda
    vect_A.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'1-DOR'].sum())
    vect_B.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'2-DOR'].sum())
    vect_C.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'3-DOR'].sum())
    vect_D.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'4-DOR'].sum())
    vect_E.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Mas de 4 dor'].sum())
    vect_F.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'No informa Num_dormit'].sum())
    vect_G.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'0-FALL'].sum())
    vect_H.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'1-FALL'].sum())
    vect_I.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'2-FALL'].sum())
    vect_J.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Mas de 2 fall'].sum())
    vect_K.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'1-PER'].sum())
    vect_L.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'2-PER'].sum())
    vect_M.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'3-PER'].sum())
    vect_N.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'4-PER'].sum())
    vect_O.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'Mas de 4 per'].sum())


df_agrup_hog['1-DOR'] = vect_A #En el nuevo dataframe, a cada columna se le asigna el valor correspondiente
df_agrup_hog['2-DOR'] = vect_B
df_agrup_hog['3-DOR'] = vect_C
df_agrup_hog['4-DOR'] = vect_D
df_agrup_hog['Mas de 4 dor'] = vect_E
df_agrup_hog['No informa Num_dormit'] = vect_F
df_agrup_hog['0-FALL'] = vect_G
df_agrup_hog['1-FALL'] = vect_H
df_agrup_hog['2-FALL'] = vect_I
df_agrup_hog['Mas de 2 fall'] = vect_J
df_agrup_hog['1-PER'] = vect_K
df_agrup_hog['2-PER'] = vect_L
df_agrup_hog['3-PER'] = vect_M
df_agrup_hog['4-PER'] = vect_N
df_agrup_hog['Mas de 4 per'] = vect_O
df_agrup_hog
```
