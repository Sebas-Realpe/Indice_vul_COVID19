```python
#Se carga cada dataset generado de personas
dataset_dir = 'TG2022_Datos\Departamentos_CSV\Pre_Departamentos_personas_procesado_15.csv'
df_d_6 = pd.read_csv(dataset_dir, sep=(',')) #Se lee el archivo csv el cual está separado por comas
df_d_6
```

```python
#Se unen todos los datasets para tener uno con toda la información
df_dataset=df_d_1.append(df_d_2 , ignore_index=True) 
df_dataset=df_dataset.append(df_d_3 , ignore_index=True)
df_dataset=df_dataset.append(df_d_4 , ignore_index=True)
df_dataset=df_dataset.append(df_d_5 , ignore_index=True)
df_dataset=df_dataset.append(df_d_6 , ignore_index=True)
df_agrup_pers = df_dataset

print(np.shape(df_dataset['U_MPIO'].unique()))
df_dataset
```

```python
#El propósito de las siguientes líneas, es revisar cuántos municipios hay.
df_agrup_pers = df_dataset.drop_duplicates(subset=['U_DPTO','U_MPIO'],keep='last', inplace=False) #Se borran todas las filas
# en las cuales los municipios están duplicados
df_agrup_pers.reset_index(drop=True, inplace=True)#Se reinicia el index, con el fin de que el número de cada fila quede ordenado

df_agrup_pers#Se imprime el dataset
```

```python
#Se exporta el dataset creado el cual cuenta con toda la información en un solo dataset
export_dir = 'TG2022_Datos/Departamentos_CSV/Pre_Departamentos_per_unido.csv' #Se crea la dirección y el nombre para guardar el dataset
df_agrup_pers.to_csv(export_dir, index=False, header=True) #Se almacena el dataset creado
```
