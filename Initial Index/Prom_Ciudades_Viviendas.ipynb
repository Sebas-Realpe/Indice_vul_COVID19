```python
TIP_VIV = {'CASA':1,'APTO':2,'TIP_CUARTO':3,'VIV_TRAD_INDIG':4,'VIV_TRAD_ETNICA':5,'OTRA_VIV':6} #Se crea un diccionario con los
#Tipos de cuartos a los que hace referencia cada una de las opciones

for vivienda,num_viv in TIP_VIV.items(): #Se realiza un ciclo for, el cual tiene en cuenta la vivienda y el número para cada ítem del diccionario
    empty_array = np.empty([len(df_dataset),1]) #Se crea una matriz vacía, de la dimensión del dataset x 1
    empty_array[:] = np.nan #Asignar todo el arreglo como nulo
    df_dataset[vivienda] = empty_array #Se agregan las nuevas columnas del diccionario TIP_VIV
    s_aux = (df_dataset.V_TIPO_VIV == num_viv).astype(int)#Teniendo en cuenta el valor de la variable V_TIPO_VIV del dataset
    #se realiza una comparación con el valor de num_viv, si este valor coincide se guarda un 1, en caso de que no coincida
    #Se guarda un cero. Es decir, es en esta variable donde se guarda la posición de la fila en la cual en el dataset original
    #en la columna V_TIPO_VIV es igual a num_viv (1,2,3,4,5 o 6).
    df_dataset[vivienda][s_aux.index] = s_aux.values #En la respectiva columna de vivienda del dataset (casa, apto, ...)
    #en cada una de las filas guardadas en s_aux se asigna el valor correspondiente; 0 en caso de que en esa fila no haya 
    #cumplido la comparación de la sentencia anterior, y 1 en caso de que el valor en que se encuentra vivienda (ej:casa)
    #en esa fila haya sido 1.0 en el dataset original, por lo que, se asigna un 1 en la respectiva fila, haciendo referencia
    #a que la comparación fue verdadera. 
    
print(s_aux.unique()) #Se verifica que aux solo contenía 0 y 1.
df_dataset #Se imprime el dataset
```

```python
#El propósito de las siguientes líneas, es revisar cuántos municipios hay.
df_agrup = df_dataset.drop_duplicates(subset=['U_DPTO','U_MPIO'],keep='last', inplace=False) #Se borran todas las filas
#En las cuales los municipios están duplicados
df_agrup.reset_index(drop=True, inplace=True)#Se reinicia el index, con el fin de que el número de cada fila quede ordenado
df_agrup#Se imprime el dataset
```

```python
val_viv_mun = df_agrup['U_MPIO'] #Se guardan los valores de la columna U_MPIO
val_viv_dpto = df_agrup['U_DPTO'] #Se guardan los valores de la columna U_DPTO
vect_A = [] #Se crea un vector vacío
vect_B = [] #Se crea un vector vacío
vect_C = [] #Se crea un vector vacío
vect_D = [] #Se crea un vector vacío
vect_E = [] #Se crea un vector vacío
vect_F = [] #Se crea un vector vacío
for n_MPIO,n_DPTO in zip(val_viv_mun,val_viv_dpto): #El ZIP se usa para empezar a asociar cada departamento con el respectivo municipio
    #Se realiza la comparacion entre el número de municipio y el de departamento, en el caso de que sean iguales
    #Se van sumando y se van guardando en un vector según el tipo de vivienda
    vect_A.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'CASA'].sum())
    vect_B.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'APTO'].sum())
    vect_C.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'TIP_CUARTO'].sum())
    vect_D.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'VIV_TRAD_INDIG'].sum())
    vect_E.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'VIV_TRAD_ETNICA'].sum())
    vect_F.append(df_dataset.loc[(df_dataset['U_MPIO'] == n_MPIO) & (df_dataset['U_DPTO'] == n_DPTO) ,'OTRA_VIV'].sum())


df_agrup['CASA'] = vect_A #En el nuevo dataframe, a cada columna se le asigna el total de cada tipo de vivienda
df_agrup['APTO'] = vect_B
df_agrup['TIP_CUARTO'] = vect_C
df_agrup['VIV_TRAD_INDIG'] = vect_D
df_agrup['VIV_TRAD_ETNICA'] = vect_E
df_agrup['OTRA_VIV'] = vect_F
df_agrup
```

```python
export_dir = 'TG2022_Datos/Departamentos_CSV/Pre_Departamentos_viv_2.csv'
df_agrup.to_csv(export_dir, index=False, header=True)
```
